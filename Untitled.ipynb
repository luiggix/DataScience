```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi as Ï€
from math import log as ln

def calc_tiempo(M, To, c = 3.7, ðœŒ = 1.038, K = 5.4e-3, Tw = 100, Ty = 70):

    # Se calcula el tiempo de cocciÃ³n 
    t = M**(2/3) * c * ðœŒ**(1/3) / ( K * Ï€**2 * (4 * Ï€ / 3)**(2/3) ) * ln(0.76 * (To - Tw) / (Ty - Tw))

    # Se convierte el tiempo al formato min:seg
    t_min = int(t / 60)
    t_seg = int(t - t_min * 60)
    
    return t_min, t_seg
```

```python
def heatmap(data, data2, row_labels, col_labels, annot=True, 
            title = "", ax = None,
            cbar_kw = None, cbar_label = "", text_kw = None,
            text_format = "{:.2f}:{:.02f}", inv_tc=False, 
            **kwargs):
    """
    Crea un mapa de calor a partir de un conjunto de datos.
    
    Parameters
    ----------
    data: 2D `ndarray` 
        Los datos para generar el mapa de calor.
        
    row_labels: `list`
        Lista de etiquetas para el eje vertical.
        
    col_labels: `list`
        Lista de etiquetas para el eje horizontal.
    
    ax: `matplotlib.axes.Axes`
        Los ejes donde se dibujarÃ¡ el mapa de calor. Opcional.
        
    cbar_kw: `dict`
        Diccionario con los argumentos para el `colorbar`. Opcional.
        
    cbar_label: `string`
        Etiqueta para el `colorbar`. Opcional.

    text_kw: `dict`
        Diccionario con los argumentos para el `Text`. Opcional.
        
    text_format: `string`
        Cadena especificando el formato del texto para los valores numÃ©ricos. Opcional.
        
    inv_tc: `bool`
        Si es True se usa (black, white), en otro caso se usa (white,black). Opcional.
        
    **kwargs:
        Argumentos adicionales para `imshow`.
        
    Returns
    -------
        im : `AxesImage`
        cbar: `Colorbar`
        
    """
    # Si no se proporcionan los ejes de manera explÃ­cita, se generan o se obtiene los que
    # estÃ©n presentes.
    if ax == None: ax = plt.gca()
    
    ax.set_title(title, fontsize=14)
    
    # Generamos el mapa de calor
    nuevo = data*60+data2
    im = ax.imshow(nuevo, **kwargs)

    ### Decoramos el mapa de calor con informaciÃ³n relevante:

    # Ponemos las ticks en el top de la figura.
    ax.tick_params(top=False, bottom=True, labeltop=False, labelbottom=True)

    # Definimos las marcas sobre los ejes y las etiquetas
    ax.set_xticks(np.arange(len(col_labels)), labels=col_labels, rotation=0, ha="center", fontsize=10, fontweight='normal')
    ax.set_yticks(np.arange(len(row_labels)), labels=row_labels, fontsize=10, fontweight='normal')

    # Ajustamos los minor ticks para dibujar una rejilla
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=2)
    ax.tick_params(which="minor", bottom=False, left=False)

    # Quitamos el marco de los ejes (spines)
    ax.spines[:].set_visible(False)    
        
    # Calculamos un umbral. 
    # * Usamos "im.norm" para normalizar los datos al intervalo [0,1]
    threshold  = im.norm(nuevo.max())/2

    # Argumentos para el texto
    if text_kw == None: text_kw = {}
    text_kw.update(ha="center", va = "center")
    
    if annot:
        # Colores para el texto        
        text_colors  = ("w", "k") if inv_tc else ("k", "w") # (k,w)
    
        # Visualizamos la informaciÃ³n numÃ©rica.
        for i in range(len(row_labels)):
            for j in range(len(col_labels)):
                # Normalizamos el dato a [0,1]
                # Comparamos con el umbral, si es mayor usamos blanco para el texto, 
                # en otro caso se usa negro. Actualizamos el diccionario de parÃ¡metros.
                
                text_kw.update(color=text_colors[int(im.norm(nuevo[i,j]) > threshold)])

                # Dibujamos el texto usando el formato "text_format".
                text = ax.text(j, i, text_format.format(data[i, j], data2[i,j]), **text_kw)    

    # Barra de color
    if cbar_kw == None: cbar_kw = {}

    
#    from mpl_toolkits.axes_grid1 import make_axes_locatable
#    divider = make_axes_locatable(ax)
#    cax = divider.append_axes("right", "5%", pad="3%")
    
#    cbar = plt.colorbar(im, cax=cax, **cbar_kw)
#    cbar.ax.set_ylabel(cbar_label, rotation=-90, va="bottom")
    plt.tight_layout()

    return im#, cbar
```

```python
X1 = 48
X2 = 61
X = list(range(X1,X2,2))
Y1 = 4
Y2 = 22
Y = list(range(Y1,Y2,2))
print(len(X), len(Y))
print(X, Y)
minutos = np.zeros((len(X), len(Y)))
segundos = np.zeros((len(X), len(Y)))

for i, M in enumerate(X):
    for j, To in enumerate(Y):
        minutos[i,j], segundos[i,j] = calc_tiempo(M, To)
```

```python
masas = [str(m) for m in X]
temps = [str(t) for t in Y]
text_kw = dict(fontsize=10, fontweight='bold')

title = 'Tiempo de cocciÃ³n de un huevo duro'
plt.figure()#figsize = (9,9))
plt.xlabel('Temperatura inicial [$^oC$]', fontsize=10, color='dimgray')
plt.ylabel('Peso [g]', fontsize=10, color='dimgray')

heatmap(minutos, segundos, masas, temps, title=title, text_format = "{:1.0f}:{:02.0f}",
        text_kw = text_kw, inv_tc=False, cmap='YlGn')

plt.savefig('huevoduro.png', dpi=300)
```
